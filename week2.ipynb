```python
set(df1.columns).intersection(df2.columns)
```

```python
merged = pd.concat([df1, df2], ignore_index=True, sort=False)
```

```python
numeric_cols = merged.select_dtypes(include=['number']).columns
merged[numeric_cols].agg(['mean','median','std']).transpose()
# mode:
{c: merged[c].mode().tolist() for c in numeric_cols}
# correlation
merged[numeric_cols].corr()
```
